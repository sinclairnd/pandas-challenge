```python
"""
The below each broken by gender
Purchase Count
Average Purchase Price
Total Purchase Value
Average Purchase Total per Person by Gender
"""

import pandas as pd
df = pd.read_csv("purchase_data.csv")

# Males

male_df = df[df["Gender"] == "Male"]
female_df = df[df['Gender'] == 'Female']

male_purchases = len(male_df)
female_purchases = len(female_df)

print("Purchase count for male customers: " +  str(male_purchases))
print("Purchase count for female customers: " + str(female_purchases))

average_male_purchase = sum(male_df['Price']) / male_purchases
average_female_purchase = sum(female_df['Price']) / female_purchases

print("Average purchase price for male customers: $" +  str(average_male_purchase))
print("Average purchase price for female customers: $" + str(average_female_purchase))


total_male_purchase = sum(male_df['Price'])
total_female_purchase = sum(female_df['Price'])

print("Total purchase price for male customers: $" + str(total_male_purchase))
print("Total purchase price for female customers: $" +  str(total_female_purchase))


males = []

for sn in set(male_df['SN']):
    #print(sn)
    df_sn = male_df[male_df['SN'] == sn]
    price_sum = sum(df_sn['Price'])
    num_purchases = len(df_sn)
    average_sn = price_sum / num_purchases
    males.append((sn, average_sn))

females = []

for sn in set(female_df['SN']):
    #print(sn)
    df_sn = female_df[female_df['SN'] == sn]
    price_sum = sum(df_sn['Price']) 
    num_purchases = len(df_sn) 
    average_sn = price_sum / num_purchases
    females.append((sn, average_sn)) 
    


print("First 5 average purchases per male customers:")
print(males[0:5])
print()
print("First 5 average purchases per female customers:")
print(females[0:5])
```

```python
def calc_stats(df, bin_num):
    
    # bin == 0 --> age group: < 10
    # bin == 1 --> age group: 10-14
    # bin == 2 --> age group: 15-19
    # bin == 3 --> age group: > 19
    
    if bin_num == 0:
        print("STATISTICS FOR AGE GROUP 1: AGE < 10")
        df = df[df['Age'] < 10]
    elif bin_num == 1:
        print("STATISTICS FOR AGE GROUP 2: AGE BETWEEN 10-14")
        df = df[(df['Age'] < 14) & (df['Age'] >= 10)]
    elif bin_num == 2:
        print("STATISTICS FOR AGE GROUP 3: AGE BETWEEN 14-19")
        df = df[(df['Age'] < 19) & (df['Age'] >= 14)]
    elif bin_num == 3:
        print("STATISTICS FOR AGE GROUP 4: AGE > 19")
        df = df[df['Age'] > 19]
    
    male_df = df[df["Gender"] == "Male"]
    female_df = df[df['Gender'] == 'Female']

    male_purchases = len(male_df)
    female_purchases = len(female_df)

    print("Purchase count for male customers: " +  str(male_purchases))
    print("Purchase count for female customers: " + str(female_purchases))

    average_male_purchase = sum(male_df['Price']) / male_purchases
    average_female_purchase = sum(female_df['Price']) / female_purchases

    print("Average purchase price for male customers: $" +  str(average_male_purchase))
    print("Average purchase price for female customers: $" + str(average_female_purchase))


    total_male_purchase = sum(male_df['Price'])
    total_female_purchase = sum(female_df['Price'])

    print("Total purchase price for male customers: $" + str(total_male_purchase))
    print("Total purchase price for female customers: $" +  str(total_female_purchase))


    males = [] # 

    for sn in set(male_df['SN']):
        #print(sn)
        df_sn = male_df[male_df['SN'] == sn]
        price_sum = sum(df_sn['Price'])
        num_purchases = len(df_sn)
        average_sn = price_sum / num_purchases
        males.append((sn, average_sn))

    females = []

    for sn in set(female_df['SN']):
        
        df_sn = female_df[female_df['SN'] == sn]
        price_sum = sum(df_sn['Price'])
        num_purchases = len(df_sn)
        average_sn = price_sum / num_purchases
        females.append((sn, average_sn))
    print()
```

```python
"""
The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
Purchase Count
Average Purchase Price
Total Purchase Value
Average Purchase Total per Person by Age Group
"""
calc_stats(df, 0)
calc_stats(df, 1)
calc_stats(df, 2)
calc_stats(df, 3)
```

```python
"""
Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
SN
Purchase Count
Average Purchase Price
Total Purchase Value
"""

users = []

for sn in set(df['SN']):

    df_sn = df[df['SN'] == sn] 
    price_sum = sum(df_sn['Price'])
    num_purchases = len(df_sn)
    average_sn = price_sum / num_purchases
    users.append((sn, num_purchases, average_sn, price_sum))

users.sort(key=lambda a: a[3], reverse = True)


top_5 = users[0:5]

df_top_5 = pd.DataFrame(top_5, columns = ['SN', 'Purchase Count', 'Average Purchase Price', 'Total Purchase Value'])

```

```python
"""
Identify the 5 most popular items by purchase count, then list (in a table):
Item ID
Item Name
Purchase Count
Item Price
Total Purchase Value
"""

items = []

for item_name in set(df['Item Name']):

    df_item = df[df['Item Name'] == item_name]
    item_price = list(df_item['Price'])[0]
    item_ID = list(df_item['Item Name'])[0]
    total_purchase_value = sum(df_item['Price'])
    num_purchases = len(df_item) 
    items.append((item_ID, item_name, num_purchases, item_price, total_purchase_value))

items.sort(key=lambda a: a[2], reverse = True)

top_5_items = items[0:5]

df_top_5_items = pd.DataFrame(top_5_items, columns = ['Item ID', 'Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value'])
df_top_5_items
```
